## Recommendation System untuk Used Car Dataset 

```python
import pandas as pd
```

```python
df = pd.read_csv('used_car_data.csv')
```

```python
data = df.copy()
```

```python
data
```

```python
a=[]
c =[]
for i in data['Name']:
    b = i.split(" ")
    a.append(b)
for j in a:
    c.append(j[0])
```

```python
data['Merk'] = c
```

```python
data1 = data.copy()
```

```python
data1
```

```python
## Perumpamaan Pembagian Tahun Mobil

tah = pd.qcut(data["Year"], 3).values.categories
tah
```

### Jadi :
- untuk tahun dibawah 2012 masuk kategori : lama
- untuk tahun dibawah 2012 masuk kategori : sedang
- untuk tahun dibawah 2012 masuk kategori : baru

```python
## Fungsi 

def tahun(x):
    if x < 2012:
        return 'lama'
    elif x > 2012 & x< 2015:
        return 'sedang'
    else:
        return 'baru'
```

```python
data1['tahun']= data['Year'].apply(tahun)
```

```python
data1
```

```python
data1['Year'] = data1['Year'].astype(str)
```

```python
data1['Recom'] = data1.apply(lambda row:str(row['Merk']) + ' ' + str(row['Fuel_Type'])+ ' ' + str(row['Transmission'])+ ' ' + str(row['Owner_Type'])+ ' ' + str(row['tahun']), axis=1)
```

```python
data1
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cv = CountVectorizer()
cm = cv.fit_transform(data1['Recom'])

# total kata unik di col B&C
print(cv.get_feature_names())

# matrix freq tiap kata di tiap data
print(cm.toarray())
```

```python
cosScore = cosine_similarity(cm.toarray())
cosScore
```

```python
### Apabila customer suka dengan mobil index ke 2 dengan beberapa feature yaitu 
## 1. Merk : Honda, 
## 2. Fuel_Type = Petrol
## 3. Transmision = Manual
## 4. Owner_Type = First
## 5. tahun = masuk kategori lama

Cust = 2
```

```python
## Detail index ke = 2

data1.iloc[Cust]
```

```python
## Pembuatan Recommendation System

data1
```

```python
mobil = list(enumerate(cosScore[Cust]))
mobil
```

```python
mobil = sorted(mobil, key=lambda x: x[1], reverse=True)
mobil
```

```python
## Berikut ini merupakan 5 mobil rekomendasi bagi customer tersebut

a = []
for i in mobil[:6]:
    a.append(data1.iloc[i[0]])
a = pd.DataFrame(a)
a = a.drop(Cust)
a
```
